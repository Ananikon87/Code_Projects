# Building a Calculator

```python
from tkmacosx import Button
```

```python
# Colors used for the calculator

co1 = "#feffff"  # white
co2 = "#6f9fbd"  # blue
co3 = "#38576b"  # value display

background = "#3b3b3b"
co10 = "#ECEFF1"

color1 = "#FFAB40"
color2 = "#ff333a"
color3 = "#6bd66f"
color4 = "#ab8918"

window = Tk()
window.title('Calculator')
window.geometry('235x318')
window.config(bg=co1)

# Top frame split for number display
frame_display = Frame(window, width=300, height=56, bg=co3)
frame_display.grid(row=0, column=0, sticky=NW)

# Bottom frame split for buttons
frame_buttons = Frame(window, width=300, height=340, bg=background)
frame_buttons.grid(row=1, column=0, sticky=NW)

# Defining functions to display selected buttons
def entering_values(event):
    global all_values

    all_values += str(event)
    value_text.set(all_values)

def calculate():
    global all_values
    try:
        if '%' in all_values:
            # Handle percentage calculation
            parts = all_values.split('%')
            if len(parts) == 2:
                a = float(parts[0])
                b = float(parts[1])
                result = a * (b / 100)  # Calculate percentage
            else:
                raise ValueError("Invalid percentage operation")
        else:
            # For other operations, use eval as before
            result = eval(all_values)
        
        # Format the result to show decimals when necessary
        if isinstance(result, float):
            result = f"{result:.6f}".rstrip('0').rstrip('.')
        else:
            result = str(result)
        
        value_text.set(result)
        all_values = ""
    except Exception as e:
        value_text.set("Error")
        all_values = ""



def clear_screen():
    global all_values
    all_values=""
    value_text.set("")
    
all_values = ""
value_text = StringVar()

# Create the display label
display = Label(frame_display, textvariable=value_text, 
                               width=20, height=3, padx=7, 
                               relief='flat', anchor='e', font=('Ivy', 18),
                               bg='#37474f', fg=co1)
display.place(x=0,y=0)

# Creating the 'Clear' button
Button(frame_buttons, command=clear_screen,
                      text='C', width=118, height=50,
                      bg=co10, fg=background,
                      font=('Ivy', 13, 'bold'),
                      borderless=1).place(x=0, y=0)

# Creating the '%' button
Button(frame_buttons, command=lambda:entering_values("%"),
                      text='%', width=59, height=50,
                      bg=co10, fg=background,
                      font=('Ivy', 13, 'bold'),
                      borderless=1).place(x=118, y=0)

# Creating the '/' button
Button(frame_buttons, command=lambda:entering_values("/"),
                      text='/', width=59, height=50,
                      bg=color1, fg=co1,
                      font=('Ivy', 13, 'bold'),
                      borderless=1).place(x=177, y=0)

# Creating number buttons for calculator
numbers = [ ('7',0,52), ('8',59,52), ('9',118,52),
            ('4',0,104), ('5',59,104), ('6',118,104),
            ('1',0,156), ('2',59,156), ('3',118,156),
            ('0',0,208)]

for num, x, y in numbers:
    width = 118 if num == '0' else 59  # 0 button is wider
    
    Button(frame_buttons, command=lambda n=num: entering_values(n),
                          text=num, width=width, height=50,
                          bg=co10, fg=background,
                          font=('Ivy', 13, 'bold'),
                          borderless=1).place(x=x, y=y)

# Creating operator buttons for calculator
operators = [('*', 177, 52),  # Multiply
             ('-', 177, 104), # Subtract
             ('+', 177, 156), # Add
             ('.', 118, 208), # Decimal point
             ('=', 177, 208)] # Equals

for op, x, y in operators:
    width = 59
    height = 50
    if op == '=':
        height = height
    
    Button(frame_buttons, command=lambda o=op: entering_values(o) if o!='=' else calculate(),
           text=op, width=width, height=height,
           bg=color1 if op != '=' else color3, fg=co1,
           font=('Ivy', 13, 'bold'),
           borderless=1).place(x=x, y=y)





window.mainloop()
```
